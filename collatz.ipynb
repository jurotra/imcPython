```python
import matplotlib.pyplot as plt
```

```python
def collatz_number(n):
    result=[n]

    while True:  # condition for stopping the function
        if n % 2==0: #even
            n = n // 2
            result.append(n)
        elif n<=1:
            break
        else:
            n = n*3+1
            result.append(n)

    return result
```

```python
n = int(input("Enter a number to use in Collatz: "))
y= collatz_number(n)
x= list(range(len(y)))
seed= n
```

```python
len(y)
```

```python
plt.figure(figsize=(8, 5))
title = "Collatz Conjecture (n=" + str(seed) + ")"
plt.title(title)
plt.plot(x,y, color="#9c9ad7", marker="o")
plt.grid(True)
plt.xlabel("Steps")
plt.ylabel("Values")
plt.yscale('log')
```
